# API Dataset, Cost of Living for each state

```python
import urllib.request
import json
from pandas.io.json import json_normalize
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go

import matplotlib.pyplot as plt
%matplotlib inline
from plotly.offline import *
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode( connected = True )
```

```python
with urllib.request.urlopen("https://www.numbeo.com/api/rankings_by_city_historical?api_key=aml7iv7ua8pnxo&section=1") as url:
    data = json.load(url)
    print(data)
```

```python
# 6 --> FROM 2012 TO 2017 DataSets
CostOfLivingCities2012 = json_normalize(data['2012'])
CostOfLivingCities2013 = json_normalize(data['2013'])
CostOfLivingCities2014 = json_normalize(data['2014'])
CostOfLivingCities2015 = json_normalize(data['2015-mid'])
CostOfLivingCities2016 = json_normalize(data['2016'])
CostOfLivingCities2017 = json_normalize(data['2017'])
print(CostOfLivingCities2016.shape)
# Look at the first three columns
CostOfLivingCities2015.head(5)
```

```python
CostOfLivingCities2012isUSA = CostOfLivingCities2012['country']=='United States'
CostOfLivingCities2013isUSA = CostOfLivingCities2013['country']=='United States'
CostOfLivingCities2014isUSA = CostOfLivingCities2014['country']=='United States'
CostOfLivingCities2015isUSA = CostOfLivingCities2015['country']=='United States'
CostOfLivingCities2016isUSA = CostOfLivingCities2016['country']=='United States'
CostOfLivingCities2017isUSA = CostOfLivingCities2017['country']=='United States'


# CostOfLivingCities2016isUSA is a boolean variable with True or False in it
CostOfLivingCities2016isUSA.head(6)
```

```python
# filter rows for Country = 'United States' using  the boolean variable
CostOfLivingCities2012USA = CostOfLivingCities2012[CostOfLivingCities2012isUSA]
CostOfLivingCities2013USA = CostOfLivingCities2013[CostOfLivingCities2013isUSA]
CostOfLivingCities2014USA = CostOfLivingCities2014[CostOfLivingCities2014isUSA]
CostOfLivingCities2015USA = CostOfLivingCities2015[CostOfLivingCities2015isUSA]
CostOfLivingCities2016USA = CostOfLivingCities2016[CostOfLivingCities2016isUSA]
CostOfLivingCities2017USA = CostOfLivingCities2017[CostOfLivingCities2017isUSA]

print(CostOfLivingCities2015USA.shape)
```

```python
#Clean DataSets:: delete unneeded columns

CostOfLivingCities2015USA.drop(['restaurant_price_index','rent_index','purchasing_power_incl_rent_index','groceries_index','cpi_and_rent_index','city_id','country'],axis=1, inplace=True )
CostOfLivingCities2015USARentIndex.drop(['cpi_index','restaurant_price_index','purchasing_power_incl_rent_index','groceries_index','cpi_and_rent_index','city_id','country'],axis=1 , inplace=True)
```

```python
# Removing the name of the city, keeping only the state abbrev 
## because we will be choosing the city (state) that has the highest Cost Of Living to represent the state
CostOfLivingCities2015USA['city_name'] = CostOfLivingCities2015USA['city_name'].apply(lambda x: x.split(',')[-1])
CostOfLivingCities2015USA['city_name'] = CostOfLivingCities2015USA['city_name'].str.strip()


CostOfLivingCities2015USARentIndex['city_name'] = CostOfLivingCities2015USARentIndex['city_name'].apply(lambda x: x.split(',')[-1])
CostOfLivingCities2015USARentIndex['city_name'] = CostOfLivingCities2015USARentIndex['city_name'].str.strip()

CostOfLivingCities2015USA.head(6)
```

```python
CostOfLivingCities2015USACopy1 = CostOfLivingCities2015USA.copy()
CostOfLivingCities2015USACopy2R = CostOfLivingCities2015USARentIndex.copy()
```

```python
CostOfLivingCities2015USACopy1.info()
```

```python
# renaming the columns (Easier to understand)
CostOfLivingCities2015USACopy1.columns = ['StateAbbrev','CostOfLiving']

CostOfLivingCities2015USACopy2R.columns = ['StateAbbrev','RentIndex']

CostOfLivingCities2015USACopy2R.info()
```

```python

# Selecting states with highest 'cpi_index=CostOfLiving' GroupBy CityName & nLargest Cpi_Index
CostOfLivingCities2015USACopy1 = CostOfLivingCities2015USACopy1.groupby('StateAbbrev')['CostOfLiving'].apply(lambda grp: grp.nlargest(1)).reset_index()
# Sorting the values (cost of living) from largest to smallest
CostOfLivingCities2015USACopy1 = CostOfLivingCities2015USACopy1.sort_values('CostOfLiving', ascending=False)[['CostOfLiving', 'StateAbbrev']].reset_index()
CostOfLivingCities2015USACopy1.drop(['index'],axis=1, inplace=True )
```

```python
# Selecting states with highest 'Rent_index=CostOfLiving' GroupBy CityName & nLargest Rent_Index
CostOfLivingCities2015USACopy2R = CostOfLivingCities2015USACopy2R.groupby('StateAbbrev')['RentIndex'].apply(lambda grp: grp.nlargest(1)).reset_index()
# Sorting the values (cost of living) from largest to smallest
CostOfLivingCities2015USACopy2R = CostOfLivingCities2015USACopy2R.sort_values('RentIndex', ascending=False)[['RentIndex', 'StateAbbrev']].reset_index()
CostOfLivingCities2015USACopy2R.drop(['index'],axis=1, inplace=True )
```

```python
CostOfLivingCities2015USACopy1.head()
```

```python

##The names column in StateAbbrev has extra white-space surrounding the text.
CostOfLivingCities2015USACopy1['StateAbbrev'] = CostOfLivingCities2015USACopy1['StateAbbrev'].astype(str).str.strip()
CostOfLivingCities2015USACopy1['State'] = CostOfLivingCities2015USACopy1['StateAbbrev']


##The names column in StateAbbrev has extra white-space surrounding the text.
CostOfLivingCities2015USACopy2R['StateAbbrev'] = CostOfLivingCities2015USACopy2R['StateAbbrev'].astype(str).str.strip()

CostOfLivingCities2015USACopy1.head()
```

```python
CostOfLivingCities2015USACopy3 = CostOfLivingCities2015USACopy1.copy()

CostOfLivingCities2015USACopy3 = CostOfLivingCities2015USACopy3.merge(CostOfLivingCities2015USACopy2R, on='StateAbbrev')
CostOfLivingCities2015USACopy3.head()
```

```python

trace0 = go.Bar(
    x=CostOfLivingCities2015USACopy3['StateAbbrev'],
    y=CostOfLivingCities2015USACopy3['CostOfLiving'],
    name='Cost of living',
    marker=dict(
        color='rgba(100,149,237, 0.85)'
    )
)
trace1 = go.Bar(
    x=CostOfLivingCities2015USACopy3['StateAbbrev'],
    y=CostOfLivingCities2015USACopy3['RentIndex'],
    name='Rent Index',
    marker=dict(
        color='rgba(204, 204, 204, 0.95)',
    )
)

data = [trace0, trace1]
layout = go.Layout(
    
    title="Cost of living & Rent Indices for each USA State, 2015",
      xaxis=dict(
         title='States Abbrevation',
        titlefont=dict(
            size=16,
            color='rgb(107, 107, 107)'
        ),
          tickangle=-45,
        tickfont=dict(
            size=12,
            color='rgb(107, 107, 107)'
        )
    ),
#    yaxis=dict(range=[0, 120]),
    
    barmode='group',
)


fig = go.Figure(data=data, layout=layout)
iplot(fig)

# https://plot.ly/python/bar-charts/
```
